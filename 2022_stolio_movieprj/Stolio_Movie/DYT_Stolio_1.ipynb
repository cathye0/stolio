```python
# 1. Linear Regression
# 공부 시간에 따른 시험 점수
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset =  pd.read_csv('/opt/google/drive/LinearRegressionData.csv')   # csv ==> 컴마로 이루어진 파일
```

```python
dataset.head()                # .head() ==> 전체 데이터를 보고 맨 위 5개 데이터를 출력
```

```python
# iloc[행(hour)(score),열(0)(1)(2)] ==> 원하는 값을 가져올 수 있는 함수
input_data=dataset.iloc[:,:-1].values # 처음부터 마지막 컬럼 직전까지의 데이터
target_data=dataset.iloc[:,-1].values  # 마지막 컬럼 데이터 (종속변수 - 결과)
```

```python
input_data,target_data
```

```python
from sklearn.linear_model import LinearRegression  # sklearn.Linear_model 패키지에서 LinearRegression 가져오기
reg = LinearRegression() # 객체생성
reg.fit(input_data,target_data) # 학습 (모델생성) 정해진 모양의 함수
```

```python
target_pred=reg.predict(input_data) # X에 대한 예측값 (리스트형태)
target_pred
```

```python
plt.scatter(input_data,target_data,color='orange') #산점도
plt.plot(input_data,target_pred,color='green') # 선 그래프
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show()
```

```python
print('9시간 공부했을 때 예상 점수 : ',reg.predict([[9]])) # 2차원이니까 대괄호 2번
                                         # input값이 2차원인 것은 원래 정해진 것
```

```python
reg.coef_ # 기울기
```

```python
reg.intercept_ # y절편
```

```python
# 데이터 세트 분리
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset =  pd.read_csv('/opt/google/drive/LinearRegressionData.csv')
dataset
```

```python
input_data = dataset.iloc[:,:-1].values
target_data = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split   # 훈련데이터와 테스트데이터를 분리 (값이 4개로 쪼개진다 (x,y,x,y))
input_train, input_test, target_train, target_test = train_test_split(input_data, target_data, test_size=0.2, random_state=0) # 훈련 80 , 테스트 20 으로 분리
# test_size : 테스트의 크기, random_state : 강의와 같은 값을 위해 랜덤하게 테스트 20 의 데이터 뽑기 
```

```python
input_data,len(input_data) #전체데이터 input_data, 개수
```

```python
input_train, len(input_train)
```

```python
input_test,len(input_test) # 테스트 세트 input_test, 개수
```

```python
target_data,len(target_data) # 전체 데이터 target_data
```

```python
target_train, len(target_train) # 훈련 세트 target_train, 개수
```

```python
target_test, len(target_test) # 테스트 세트 target_test, 개수
```

```python

# 분리된 데이터를 통한 모델링
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() # 회귀모델을 만들기 위한 객체 생성
```

```python
reg.fit(input_train,target_train) # 훈련 세트로 학습
```

```python
# 데이터 시각화(훈련 세트)
```

```python
plt.scatter(input_train,target_train,color='orange') #산점도
plt.plot(input_train, reg.predict(input_train), color='green') # 선 그래프
plt.title('Score by hours (train data)') # 제목
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show()
```

```python
# 데이터 시각화 (테스트 세트)
```

```python
plt.scatter(input_test,target_test,color='orange') #산점도
plt.plot(input_train, reg.predict(input_train), color='green') # 선 그래프 (모델을 만들 때에는 훈련세트를 사용)
plt.title('Score by hours (train data)') # 제목
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# 모델평가
```

```python
reg.score(input_test, target_test) # 테스트 세트를 통한 모델 평가
```

```python
reg.score(input_train, target_train) # 훈련 세트를 통한 모델 평가
```

```python
# 경사하강법 (Gradiant Descent)
```

```python
from sklearn.linear_model import SGDRegressor # SDG : Stochastic Gradient Descent 확률적 경사 하강법

# 지수표기법
# 1e-3 : 0.001 (10^-3)
# 1e-4 :0.001 (10^-4)
# 1e+3 : 1000 (10^3)
# 1e+4 : 10000 (10^4)

# sr = SGDRegressor(max_iter=200, eta0=1e-4, random_state=0, verbose=1)   
# sr : 객체, max_iter: 에포크 반복 횟수(작으면 m값으로 충분히 수렴이 어려울 지도 몰라) 값이 변하지 않는다고 판단하면 일찍 종료 할수도.....
# , eta0: 학습률(줄이면 에포크가 커진다), verbose: 훈련 반복 시 loss가 어떻게 줄어드는 지 보여준다
sr = SGDRegressor() # 비워두면 기본 값
sr.fit(input_train,target_train) # 모델학습
```

```python
plt.scatter(input_train,target_train,color='orange') #산점도
plt.plot(input_train, sr.predict(input_train), color='green') # 선 그래프 (모델을 만들 때에는 훈련세트를 사용)
plt.title('Score by hours (train data, SGD)') # 제목
plt.xlabel('hours') # x축 이름
plt.ylabel('score') # y축 이름
plt.show()                  # 경사하강법으로 기울기를 찾았습니다!
```

```python
sr.coef_, sr.intercept_ # train 데이터와 비교 (sum을 최소화 하는 방법)
```

```python
sr.score(input_test, target_test) # 테스트 세트를 통한 모델 평가
```

```python
sr.score(input_train, target_train) # 훈련 세트를 통한 모델 평가
```
