```python
# 원-핫 인코딩
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Stolio_Movie/ScikitLearn/LinearRegressionData.csv')
input_data = dataset.iloc[:,:-1].values
target_data = dataset.iloc[:,-1].values     # 독립변수는 리스트 형태로 값이 3개 출력값은 하나
```

```python
input_data
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct= ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first'),[2])], remainder='passthrough') 
# (튜플형태로 값 3개), encoder, OneHotEncoder: 원핫인코딩하기
# drop: 다중 공산성 문제 해결 [2]의 인덱스에만 적용, remainder: 나머지 데이터는 passthrough 그냥 지나가기
input_data=ct.fit_transform(input_data)  # 바뀐데이터 다시 집어넣어주기
input_data

# 1 0 : Home
# 0 1 : Library
# 0 0 : Cafe
```

```python
# 데이터 세트 분리
```

```python
from sklearn.model_selection import train_test_split
input_train, input_test, target_train, target_test= train_test_split(input_data, target_data, test_size=0.2, random_state=0) # test_size=test 데이터 20
```

```python
# 학습(다중 선형 회귀)
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(input_train, target_train)
```

```python
# 예측 값과 실제 값 비교 (테스트 세트)
```

```python
target_pred= reg.predict(input_test)
target_pred
```

```python
target_test  # target_pred 와 target_test를 비교
```

```python
reg.coef_ # (변수가 4개 이므로 기울기도 4개 입니닷)
```

```python
reg.intercept_
```

```python
# 모델평가
```

```python
reg.score(input_train, target_train) # 훈련 세트
```

```python
reg.score(input_test, target_test) # 테스트 세트 
```

```python
# 다양한 평가 지표(회귀 모델)
```

```python
# 1. MAE (Mean Absolute Error) : (실제 값과 예측 값) 차이의 절대값
# 1. MSE (Mean Squarred Error) : 차이의 제곱
# 1. RMSE (Root Mean Error) : 차이의 제곱에 루트
# 1. R2 : 결정계수

# R2는 1에 가까울수록, 나머지는 0에 가까울수록 좋음
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(target_test, target_pred) # 실제 값, 예측 값 # MAE
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(target_test, target_pred) # 실제 값, 예측 값 # MSE
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(target_test, target_pred,squared=False) # 실제 값, 예측 값 # RMSE
```

```python
from sklearn .metrics import r2_score
r2_score(target_test, target_pred) # R2 // LinearRegression에서 score값을 찍으면 R2값이 나와용
```
